```
import tensorflow as tf
print(f"Tensorflow Version={tf.__version__}")

import tensorflow.keras
print(f"tensorflow.keras Version={tensorflow.keras.__version__}")

#載入Dataset
from tensorflow.keras import datasets

(X_train, y_train),(X_test, y_test)= datasets.mnist.load_data()

from tensorflow.keras import layers
from tensorflow.keras import models

#建立Model
model = models.Sequential(
    [
        ######開始training Seqential model（注意！需加Flatten Layer）######
        #此列必備！
        #若沒有加Flatten則會出現可怕錯誤：
        #InvalidArgumentError:  assertion failed: [Condition x == y did not hold element-wise:] [x (sparse_categorical_crossentropy/SparseSoftmaxCrossEntropyWithLogits/Shape_1:0) = ] [32 1] [y (sparse_categorical_crossentropy/SparseSoftmaxCrossEntropyWithLogits/strided_slice:0) = ] [32 28]
        # [[node sparse_categorical_crossentropy/SparseSoftmaxCrossEntropyWithLogits/assert_equal_1/Assert/Assert (defined at <ipython-input-1-a2c0f580ffbb>:41) ]] [Op:__inference_train_function_640]
        #Function call stack:
        #train_function
        #將input轉成一維向量
        layers.Flatten(),
        ######開始training Seqential model（注意！需加Flatten Layer）######
        #建立隱藏層（此例有256個神經元）
        layers.Dense(units=256),
        #建立輸出層（輸出神經元=10）
        layers.Dense(units=10),
    ]
)

#最佳化設定
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

######開始training Seqential model######
#開始進行training
#使用fit會自動建立模型，無須手動呼叫model.build
model.fit(
    x=X_train,
    y=y_train,
    epochs=1,
    verbose=1,
    validation_split=0.1
)

model.summary()
######開始training Seqential model######


#開始進行testing
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'test_loss={test_loss}')
print(f'test_accuracy={test_accuracy}')
```

```
import tensorflow as tf
print(f"Tensorflow Version={tf.__version__}")

import tensorflow.keras
print(f"tensorflow.keras Version={tensorflow.keras.__version__}")

#載入Dataset
from tensorflow.keras import datasets

(X_train, y_train),(X_test, y_test)= datasets.mnist.load_data()

######建立一個隱藏層######
#建立Model
from tensorflow.keras import layers
from tensorflow.keras.models import Model

class Network(Model):
    def __init__(self):
        super(Network,self).__init__()
        self.flatten = layers.Flatten()
        self.layer_1 = layers.Dense(60 , activation='relu')
        self.output_layer = layers.Dense(10 , activation='softmax')
        
    def call(self,x):
        x = self.flatten(x)
        x = self.layer_1(x)
        predict = self.output_layer(x)
        return predict
model = Network()
######建立一個隱藏層######

#最佳化設定
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
#開始進行training
model.fit(
    x=X_train,
    y=y_train,
    epochs=1,
    verbose=1,
    validation_split=0.1
)
model.summary()
#開始進行testing
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'test_loss={test_loss}')
print(f'test_accuracy={test_accuracy}')
```

##設定是否打亂（Shuffle）training dataset

##監控訓練狀況（手動繪出training狀況圖：training accuracy＆validation accuracy）

```
import tensorflow as tf
print(f"Tensorflow Version={tf.__version__}")

import tensorflow.keras
print(f"tensorflow.keras Version={tensorflow.keras.__version__}")

#載入Dataset
from tensorflow.keras import datasets

(X_train, y_train),(X_test, y_test)= datasets.mnist.load_data()
num_classes =10

from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)

from tensorflow.keras import layers
from tensorflow.keras import models
model = models.Sequential(
    [
        layers.Flatten(),                
        layers.Dense(units=20,activation="relu"),
        #建立輸出層（輸出神經元個數=10）                
        layers.Dense(units=10,activation='softmax'),
    ]
) 
#最佳化設定
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#開始進行training

######監控訓練狀況（取得繪圖資訊）（1/2）######
training_history = model.fit(
######監控訓練狀況（取得繪圖資訊）（1/2）######    
    x=X_train,
    y=y_train,
    epochs=10,
    verbose=1,
    validation_split=0.1
)
model.summary()
#開始進行testing
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'test_loss={test_loss}')
print(f'test_accuracy={test_accuracy}')

######監控訓練狀況（取得繪圖資訊）（2/2）######
import matplotlib.pyplot as plt  
def plot_train_history():
    #設定x軸名稱  
    plt.xlabel('epoch')  
    #設定y軸名稱
    plt.ylabel('Accuracy')
    #開始繪製training Accuracy曲線
    plt.plot(training_history.history['accuracy'])  
    #開始繪製validation Accuracy曲線
    plt.plot(training_history.history['val_accuracy'])  
    #在圖表左上方放置legend圖（也就是用來指名哪個顏色的線條是何功用）
    plt.legend(['train', 'validation'], loc='upper left')  
    plt.show()  

plot_train_history()      
######監控訓練狀況（取得繪圖資訊）（2/2）######
```

```
import numpy as np
a = np.array(
    [
#index:  0 1  2  3
     [0, 8, -1, 7],
     
#index:  4 5  6  7
     [2, 9, 0, -9]
    ]
)

#把所有值打散成一個向量，回傳最大值的index
#因為9為最大值，而9的index為5
#因此將輸出5
print(f'np.argmax(a)={np.argmax(a)}')

#行的比較（axis=0）：
#0和2比回傳較大者的index=1
#8和9比回傳較大者的index=1
#-1和0比回傳較大者的index=1
#7和-9比回傳較大者的index=0
print(f'np.argmax(a, axis=0)={np.argmax(a, axis=0)}')

#列的比較（axis=1）：
#以第0列來說：0、8、1、7最大為8，而8的index為1
#以第1列來說：2、9、0、-9最大為9，而9的index為1
print(f'np.argmax(a, axis=1)={np.argmax(a, axis=1)}')  #橫著比較，返回列最大值的index
```

```
import tensorflow as tf
print(f"Tensorflow Version={tf.__version__}")

import tensorflow.keras
print(f"tensorflow.keras Version={tensorflow.keras.__version__}")

#載入Dataset
from tensorflow.keras import datasets

(X_train, y_train),(X_test, y_test)= datasets.mnist.load_data()

#建立Model
from tensorflow.keras import layers
from tensorflow.keras import models
model = models.Sequential(
    [
        layers.Flatten(),                
        #建立輸出層（輸出神經元個數=10）                
        layers.Dense(units=10,activation='softmax'),
    ]
) 
#最佳化設定
model.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
#開始進行training
model.fit(
    x=X_train,
    y=y_train,
    epochs=1,
    verbose=1,
    validation_split=0.1
)
#開始進行testing
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'model test_loss={test_loss}')
print(f'model test_accuracy={test_accuracy}')

######手動儲存/載入模型到json中，並另外單獨儲存權重######
#開始儲存整個模型（只含架構，不包含權重）
#model.save('latest_keras_model.h5')
json_string = model.to_json()  
with open('latest_keras_model.json','w') as json_file1:
  json_file1.write(json_string)
print('寫入模型json檔完畢（不包含權重）')

#開始儲存整個模型的權重（不包含架構本身）
model.save_weights('latest_keras_model_weight.h5')
print('寫入模型權重完畢')

#開始載入整個模型（只含架構，不包含權重）
#model2 = models.load_model('latest_keras_model.h5')
#建立Model
#model = models.Sequential()
with open('latest_keras_model.json','r') as json_file1:
  json_string = json_file1.read()

from keras.models import model_from_json
model2 = model_from_json(json_string)  
print('讀取模型json檔完畢（不包含權重）')

#開始載入整個模型的權重（不包含架構本身）
model2.load_weights('latest_keras_model_weight.h5')
print('讀取模型權重完畢')
#分開讀取模型本身和權重的話需額外compile
model2.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

######手動儲存/載入模型到json中，並另外單獨儲存權重######

#開始進行testing model2
test_loss, test_accuracy = model2.evaluate(X_test, y_test)
print(f'model2 test_loss={test_loss}')
print(f'model2 test_accuracy={test_accuracy}')
```

```
import tensorflow as tf
print(f"Tensorflow Version={tf.__version__}")

import tensorflow.keras
print(f"tensorflow.keras Version={tensorflow.keras.__version__}")

#載入Dataset
from tensorflow.keras import datasets

(X_train, y_train),(X_test, y_test)= datasets.mnist.load_data()

from tensorflow.keras import layers
from tensorflow.keras import models
model = models.Sequential(
    [
        layers.Flatten(),        
        #建立輸出層（輸出神經元個數=10）                
        layers.Dense(units=10,activation="softmax"),
    ]
) 

######自動儲存最佳模型（1/2）######
from tensorflow.keras.callbacks import ModelCheckpoint 
checkpoint = ModelCheckpoint(
    'auto_saved_best_model_improved.h5',  
    #要以loss為主或是以accuracy為主
    monitor='val_accuracy', 
    #save_best_only= True表示只存最佳模型
    save_best_only= True, 
    #若是設定auto表示系統會自動偵測應設為min或max
    mode='auto'
)  
######自動儲存最佳模型（1/2）######

#最佳化設定
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
#開始進行training
model.fit(
    x=X_train,
    y=y_train,
    epochs=100,
    verbose=1,
    validation_split=0.1,
    ######自動儲存最佳模型（2/2）######
    callbacks=[checkpoint],
    ######自動儲存最佳模型（2/2）######
)
model.summary()
#開始進行testing
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'test_loss={test_loss}')
print(f'test_accuracy={test_accuracy}')

#開始載入整個模型（含架構、權重）
model2 = models.load_model('auto_saved_best_model_improved.h5')
#開始進行testing
#會發現儲存的模型不一定真的是最佳的，但至少能自動儲存＆不會存到太爛的模型！
test_loss, test_accuracy = model2.evaluate(X_test, y_test)
print(f'test_loss={test_loss}')
print(f'test_accuracy={test_accuracy}')
```
